# 30-Minute Fallback Food Segmentation Training

This notebook is a stronger fallback when the full SegFormer-B5 run is taking too long. It keeps the same artifact format as the full pipeline, but uses SegFormer-B2 at 512px so it can train much faster while still using all FoodSeg103 training images.

1. Download FoodSeg103 from the internet through Hugging Face Datasets.
2. Train a food-specific supervised semantic segmenter using SegFormer-B2 initialized from ADE20K weights.
3. Save the full model, processor, label map, training config, and metrics.
4. Zip `best_model.zip` so it can be uploaded into the Gradio demo notebook.

Target runtime is roughly 15-35 minutes on H100 after dataset/model cache is warm. For the full SoTA-ish run, use `foodseg_hybrid_sam31_modal.ipynb` instead.

```python
import json
import math
import os
import random
import shutil
import time
from contextlib import nullcontext
from dataclasses import asdict, dataclass
from pathlib import Path

import numpy as np
import torch
from PIL import Image, ImageEnhance

os.environ.setdefault("HF_HUB_ENABLE_HF_TRANSFER", "1")
os.environ.setdefault("TOKENIZERS_PARALLELISM", "false")

ROOT = Path.cwd()
DATA_DIR = ROOT / "data"
OUTPUT_DIR = ROOT / "outputs"
CACHE_DIR = DATA_DIR / "hf_cache"
RUN_NAME = time.strftime("foodseg_30min_segformer_b2_%Y%m%d_%H%M%S")
RUN_DIR = OUTPUT_DIR / RUN_NAME

for path in [DATA_DIR, OUTPUT_DIR, CACHE_DIR, RUN_DIR]:
    path.mkdir(parents=True, exist_ok=True)

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

if torch.cuda.is_available():
    torch.backends.cuda.matmul.allow_tf32 = True
    torch.backends.cudnn.allow_tf32 = True

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
GPU_NAME = torch.cuda.get_device_name(0) if torch.cuda.is_available() else "cpu"
BF16 = torch.cuda.is_available() and torch.cuda.get_device_capability(0)[0] >= 8

print({"device": DEVICE, "gpu": GPU_NAME, "bf16": BF16, "run_dir": str(RUN_DIR)})
```

```python
@dataclass
class TrainConfig:
    dataset_id: str = "EduardoPacheco/FoodSeg103"
    model_name: str = "nvidia/segformer-b2-finetuned-ade-512-512"
    image_size: int = 512
    num_labels: int = 104
    ignore_index: int = 255
    epochs: int = 10
    train_batch_size: int = 16
    eval_batch_size: int = 8
    gradient_accumulation_steps: int = 1
    learning_rate: float = 8e-5
    weight_decay: float = 0.01
    warmup_ratio: float = 0.06
    max_train_samples: int | None = None
    max_eval_samples: int | None = 1024
    use_class_weights: bool = True
    class_weight_samples: int = 1800
    dataloader_num_workers: int = max(2, min(12, (os.cpu_count() or 8) // 2))
    save_total_limit: int = 2
    report_to: str = "tensorboard"

CFG = TrainConfig()

# Strong fallback budget. Set GPU_PRESET=b200 for a larger still-fast run.
GPU_PRESET = os.environ.get("GPU_PRESET", "h100").lower()

if GPU_PRESET == "b200":
    CFG.train_batch_size = 32
    CFG.eval_batch_size = 16
    CFG.epochs = 14
    CFG.max_train_samples = None
    CFG.max_eval_samples = 1536
    CFG.class_weight_samples = 2400
elif GPU_PRESET == "h100":
    CFG.train_batch_size = 16
    CFG.eval_batch_size = 8
    CFG.epochs = 10
    CFG.max_train_samples = None
    CFG.max_eval_samples = 1024
    CFG.class_weight_samples = 1800

with open(RUN_DIR / "train_config.json", "w", encoding="utf-8") as f:
    json.dump(asdict(CFG), f, indent=2)

print({"gpu_preset": GPU_PRESET, "mode": "30min_fallback"})
print(json.dumps(asdict(CFG), indent=2))
```

```python
import matplotlib.pyplot as plt

rng = np.random.default_rng(SEED)
PALETTE = rng.integers(0, 255, size=(CFG.num_labels, 3), dtype=np.uint8)
PALETTE[0] = np.array([0, 0, 0], dtype=np.uint8)

def mask_to_rgb(mask: np.ndarray) -> np.ndarray:
    safe = np.where((mask >= 0) & (mask < CFG.num_labels), mask, 0).astype(np.int64)
    return PALETTE[safe]

def overlay_mask(image: Image.Image, mask: np.ndarray, alpha: float = 0.48) -> Image.Image:
    image_np = np.array(image.convert("RGB"))
    color = mask_to_rgb(mask)
    foreground = mask > 0
    blended = image_np.copy()
    blended[foreground] = (image_np[foreground] * (1 - alpha) + color[foreground] * alpha).astype(np.uint8)
    return Image.fromarray(blended)

def show_example(idx: int = 0, split: str = "train"):
    ex = raw_ds[split][idx]
    image = ex["image"].convert("RGB")
    mask = np.array(ex["label"])
    labels = [id2label[int(i)] for i in np.unique(mask) if int(i) in id2label]
    fig, axes = plt.subplots(1, 3, figsize=(16, 5))
    axes[0].imshow(image)
    axes[0].set_title("image")
    axes[1].imshow(mask_to_rgb(mask))
    axes[1].set_title("mask")
    axes[2].imshow(overlay_mask(image, mask))
    axes[2].set_title(", ".join(labels[:8]))
    for ax in axes:
        ax.axis("off")
    plt.tight_layout()

show_example(0)
```

```python
from torchvision.transforms import InterpolationMode
from torchvision.transforms import functional as TVF
from transformers import SegformerImageProcessor

image_processor = SegformerImageProcessor.from_pretrained(
    CFG.model_name,
    do_resize=False,
    reduce_labels=False,
)

def sanitize_mask(mask: Image.Image) -> Image.Image:
    arr = np.array(mask)
    if arr.ndim == 3:
        arr = arr[..., 0]
    arr = arr.astype(np.int64)
    arr[(arr < 0) | (arr >= CFG.num_labels)] = CFG.ignore_index
    return Image.fromarray(arr.astype(np.uint8), mode="L")

def paired_resize(image: Image.Image, mask: Image.Image, size: int):
    image = TVF.resize(image, [size, size], interpolation=InterpolationMode.BICUBIC, antialias=True)
    mask = TVF.resize(mask, [size, size], interpolation=InterpolationMode.NEAREST)
    return image, mask

def color_jitter_light(image: Image.Image) -> Image.Image:
    if random.random() < 0.5:
        image = ImageEnhance.Brightness(image).enhance(random.uniform(0.85, 1.18))
    if random.random() < 0.5:
        image = ImageEnhance.Contrast(image).enhance(random.uniform(0.85, 1.18))
    if random.random() < 0.35:
        image = ImageEnhance.Color(image).enhance(random.uniform(0.85, 1.18))
    return image

def train_transform(batch):
    images, masks = [], []
    for image, mask in zip(batch["image"], batch["label"]):
        image = image.convert("RGB")
        mask = sanitize_mask(mask)
        if random.random() < 0.5:
            image = TVF.hflip(image)
            mask = TVF.hflip(mask)
        image = color_jitter_light(image)
        image, mask = paired_resize(image, mask, CFG.image_size)
        images.append(image)
        masks.append(np.array(mask, dtype=np.int64))
    return image_processor(images=images, segmentation_maps=masks, return_tensors="pt")

def val_transform(batch):
    images, masks = [], []
    for image, mask in zip(batch["image"], batch["label"]):
        image = image.convert("RGB")
        mask = sanitize_mask(mask)
        image, mask = paired_resize(image, mask, CFG.image_size)
        images.append(image)
        masks.append(np.array(mask, dtype=np.int64))
    return image_processor(images=images, segmentation_maps=masks, return_tensors="pt")

train_ds = raw_ds["train"].with_transform(train_transform)
eval_ds = raw_ds["validation"].with_transform(val_transform)

def collate_fn(batch):
    return {
        "pixel_values": torch.stack([item["pixel_values"] for item in batch]),
        "labels": torch.stack([item["labels"] for item in batch]).long(),
    }

batch = collate_fn([train_ds[i] for i in range(2)])
print({k: tuple(v.shape) for k, v in batch.items()}, batch["labels"].dtype)
```

```python
def compute_class_weights(dataset, max_samples: int) -> torch.Tensor:
    counts = np.zeros(CFG.num_labels, dtype=np.float64)
    n = min(max_samples, len(dataset))
    for idx in range(n):
        mask = np.array(dataset[idx]["label"])
        if mask.ndim == 3:
            mask = mask[..., 0]
        valid = (mask >= 0) & (mask < CFG.num_labels)
        ids, vals = np.unique(mask[valid], return_counts=True)
        counts[ids.astype(np.int64)] += vals
    freq = counts / max(counts.sum(), 1.0)
    weights = 1.0 / np.log(1.02 + np.maximum(freq, 1e-12))
    weights[counts == 0] = 0.0
    weights = weights / np.mean(weights[weights > 0])
    weights[0] = min(weights[0], 0.35)
    return torch.tensor(weights, dtype=torch.float32)

class_weights = None
if CFG.use_class_weights:
    class_weights = compute_class_weights(raw_ds["train"], CFG.class_weight_samples)
    torch.save(class_weights, RUN_DIR / "class_weights.pt")
    top = sorted([(float(w), id2label[i]) for i, w in enumerate(class_weights)], reverse=True)[:10]
    print("largest weights:", top)
```

```python
import torch.nn.functional as F
from transformers import AutoConfig, SegformerForSemanticSegmentation, Trainer, TrainingArguments
import inspect

model_config = AutoConfig.from_pretrained(CFG.model_name)
model_config.num_labels = CFG.num_labels
model_config.id2label = id2label
model_config.label2id = label2id

model = SegformerForSemanticSegmentation.from_pretrained(
    CFG.model_name,
    config=model_config,
    ignore_mismatched_sizes=True,
)
if getattr(model, "supports_gradient_checkpointing", False):
    model.gradient_checkpointing_enable()
    print("gradient checkpointing enabled")
else:
    print("gradient checkpointing skipped: SegFormer build does not support it")

def preprocess_logits_for_metrics(logits, labels):
    if isinstance(logits, tuple):
        logits = logits[0]
    logits = F.interpolate(logits, size=labels.shape[-2:], mode="bilinear", align_corners=False)
    return logits.argmax(dim=1)

def segmentation_confusion_matrix(preds, labels, num_labels: int, ignore_index: int):
    preds = np.asarray(preds, dtype=np.int64)
    labels = np.asarray(labels, dtype=np.int64)
    valid = labels != ignore_index
    valid &= labels >= 0
    valid &= labels < num_labels
    valid &= preds >= 0
    valid &= preds < num_labels
    if not np.any(valid):
        return np.zeros((num_labels, num_labels), dtype=np.int64)
    encoded = num_labels * labels[valid] + preds[valid]
    hist = np.bincount(encoded, minlength=num_labels * num_labels)
    return hist.reshape(num_labels, num_labels)

def compute_metrics(eval_pred):
    preds, labels = eval_pred
    hist = segmentation_confusion_matrix(preds, labels, CFG.num_labels, CFG.ignore_index)
    true_positive = np.diag(hist).astype(np.float64)
    pred_area = hist.sum(axis=0).astype(np.float64)
    label_area = hist.sum(axis=1).astype(np.float64)
    union = pred_area + label_area - true_positive

    iou = np.divide(true_positive, union, out=np.full_like(true_positive, np.nan), where=union > 0)
    accuracy = np.divide(true_positive, label_area, out=np.full_like(true_positive, np.nan), where=label_area > 0)
    present = union > 0
    present_no_bg = present.copy()
    if len(present_no_bg):
        present_no_bg[0] = False

    total = hist.sum()
    overall_accuracy = float(true_positive.sum() / total) if total > 0 else 0.0
    mean_iou = float(np.nanmean(iou[present])) if np.any(present) else 0.0
    mean_accuracy = float(np.nanmean(accuracy[label_area > 0])) if np.any(label_area > 0) else 0.0
    mean_iou_no_background = float(np.nanmean(iou[present_no_bg])) if np.any(present_no_bg) else mean_iou
    fw_iou = float(np.nansum(label_area[present] * iou[present]) / np.sum(label_area[present])) if np.any(present) else 0.0

    return {
        "mean_iou": mean_iou,
        "mean_iou_no_background": mean_iou_no_background,
        "mean_accuracy": mean_accuracy,
        "overall_accuracy": overall_accuracy,
        "frequency_weighted_iou": fw_iou,
        "num_classes_present": int(np.sum(present)),
    }

class WeightedSegformerTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = F.interpolate(outputs.logits, size=labels.shape[-2:], mode="bilinear", align_corners=False)
        weight = class_weights.to(logits.device) if class_weights is not None else None
        loss = F.cross_entropy(logits, labels, weight=weight, ignore_index=CFG.ignore_index)
        return (loss, outputs) if return_outputs else loss

args_kwargs = dict(
    output_dir=str(RUN_DIR / "trainer"),
    per_device_train_batch_size=CFG.train_batch_size,
    per_device_eval_batch_size=CFG.eval_batch_size,
    gradient_accumulation_steps=CFG.gradient_accumulation_steps,
    learning_rate=CFG.learning_rate,
    weight_decay=CFG.weight_decay,
    num_train_epochs=CFG.epochs,
    lr_scheduler_type="cosine",
    warmup_ratio=CFG.warmup_ratio,
    logging_steps=25,
    save_strategy="epoch",
    save_total_limit=CFG.save_total_limit,
    load_best_model_at_end=True,
    metric_for_best_model="mean_iou",
    greater_is_better=True,
    bf16=BF16,
    fp16=(torch.cuda.is_available() and not BF16),
    optim="adamw_torch_fused" if torch.cuda.is_available() else "adamw_torch",
    remove_unused_columns=False,
    dataloader_num_workers=CFG.dataloader_num_workers,
    eval_accumulation_steps=8,
    report_to=[CFG.report_to] if CFG.report_to else [],
    seed=SEED,
)
sig = inspect.signature(TrainingArguments.__init__)
if "eval_strategy" in sig.parameters:
    args_kwargs["eval_strategy"] = "epoch"
else:
    args_kwargs["evaluation_strategy"] = "epoch"

training_args = TrainingArguments(**args_kwargs)

trainer = WeightedSegformerTrainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=eval_ds,
    data_collator=collate_fn,
    compute_metrics=compute_metrics,
    preprocess_logits_for_metrics=preprocess_logits_for_metrics,
)

print("trainer ready")
```

```python
effective_batch = CFG.train_batch_size * CFG.gradient_accumulation_steps
steps_per_epoch = math.ceil(len(train_ds) / effective_batch)
total_steps = steps_per_epoch * CFG.epochs
print({
    "gpu": GPU_NAME,
    "preset": GPU_PRESET,
    "mode": "30min_fallback",
    "train_samples": len(train_ds),
    "eval_samples": len(eval_ds),
    "image_size": CFG.image_size,
    "train_batch_size": CFG.train_batch_size,
    "gradient_accumulation_steps": CFG.gradient_accumulation_steps,
    "effective_batch": effective_batch,
    "epochs": CFG.epochs,
    "steps_per_epoch": steps_per_epoch,
    "total_steps": total_steps,
})

model.to(DEVICE)
model.eval()
preflight_batch = collate_fn([train_ds[i] for i in range(min(2, len(train_ds)))])
preflight_batch = {k: v.to(DEVICE) for k, v in preflight_batch.items()}
with torch.inference_mode(), torch.autocast(device_type="cuda", dtype=torch.bfloat16, enabled=(DEVICE == "cuda" and BF16)):
    preflight_outputs = model(pixel_values=preflight_batch["pixel_values"])
    preflight_logits = F.interpolate(preflight_outputs.logits.float(), size=preflight_batch["labels"].shape[-2:], mode="bilinear", align_corners=False)
    preflight_preds = preflight_logits.argmax(dim=1).detach().cpu().numpy()
preflight_labels = preflight_batch["labels"].detach().cpu().numpy()
preflight_metrics = compute_metrics((preflight_preds, preflight_labels))
print({
    "preflight_pixel_values": tuple(preflight_batch["pixel_values"].shape),
    "preflight_labels": tuple(preflight_batch["labels"].shape),
    "preflight_logits": tuple(preflight_outputs.logits.shape),
    "preflight_metrics": preflight_metrics,
})
model.train()
del preflight_batch, preflight_outputs, preflight_logits, preflight_preds, preflight_labels, preflight_metrics
if torch.cuda.is_available():
    torch.cuda.empty_cache()
print("preflight ok - safe to start trainer.train()")
```

```python
# SAM still-image concept prompting. This uses the official SAM 3 image API.
# Gated checkpoints may require HF_TOKEN and accepted access on Hugging Face.
SAM3_IMAGE_MODEL = None
SAM3_IMAGE_PROCESSOR = None

def load_sam3_image_model():
    global SAM3_IMAGE_MODEL, SAM3_IMAGE_PROCESSOR
    if SAM3_IMAGE_MODEL is not None:
        return SAM3_IMAGE_MODEL, SAM3_IMAGE_PROCESSOR
    from sam3.model_builder import build_sam3_image_model
    from sam3.model.sam3_image_processor import Sam3Processor

    SAM3_IMAGE_MODEL = build_sam3_image_model(device=DEVICE, eval_mode=True, load_from_HF=True, compile=False)
    SAM3_IMAGE_MODEL = SAM3_IMAGE_MODEL.float()
    SAM3_IMAGE_PROCESSOR = Sam3Processor(SAM3_IMAGE_MODEL)
    return SAM3_IMAGE_MODEL, SAM3_IMAGE_PROCESSOR

def sam3_inference_context():
    return torch.autocast(device_type="cuda", enabled=False) if DEVICE == "cuda" else nullcontext()

def sam3_prompt_masks(image: Image.Image, prompt: str, score_threshold: float = 0.35, max_masks: int = 12):
    if not prompt or not prompt.strip():
        return [], [], []
    _, processor = load_sam3_image_model()
    with torch.inference_mode(), sam3_inference_context():
        state = processor.set_image(image.convert("RGB"))
        output = processor.set_text_prompt(state=state, prompt=prompt.strip())
    masks = output.get("masks")
    boxes = output.get("boxes")
    scores = output.get("scores")
    if masks is None:
        return [], [], []
    masks = masks.detach().float().cpu()
    if masks.ndim == 4:
        masks = masks[:, 0]
    scores_np = scores.detach().float().cpu().numpy() if scores is not None else np.ones((masks.shape[0],), dtype=np.float32)
    keep = np.where(scores_np >= score_threshold)[0]
    keep = keep[np.argsort(scores_np[keep])[::-1]][:max_masks]
    kept_masks = [(masks[i].numpy() > 0).astype(np.uint8) for i in keep]
    kept_scores = [float(scores_np[i]) for i in keep]
    kept_boxes = boxes.detach().float().cpu().numpy()[keep].tolist() if boxes is not None and len(keep) else []
    return kept_masks, kept_scores, kept_boxes

def overlay_binary_masks(image: Image.Image, masks: list[np.ndarray], scores: list[float] | None = None):
    image_np = np.array(image.convert("RGB"))
    overlay = image_np.copy()
    for i, mask in enumerate(masks):
        if mask.shape != image_np.shape[:2]:
            mask_img = Image.fromarray((mask * 255).astype(np.uint8)).resize(image.size, Image.NEAREST)
            mask = np.array(mask_img) > 0
        color = PALETTE[(i * 17 + 9) % len(PALETTE)]
        overlay[mask.astype(bool)] = (overlay[mask.astype(bool)] * 0.45 + color * 0.55).astype(np.uint8)
    return Image.fromarray(overlay)

print("SAM helper definitions are present but sam3 is not installed in this 30-minute fallback notebook.")
```
